```python
import pickle
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
with open('lsoa_metrics.pkl', 'rb') as f:
    metrics = pickle.load(f)

lsoa_list = []
lsoa_exp_list = []
lsoa_od_list = []

for k in metrics.keys():
    lsoa_list.append(metrics[k]['LSOA'])
    lsoa_exp_list.append(metrics[k]['LSOA_Expanded'])
    lsoa_od_list.append(metrics[k]['LSOA_OD'])
    
lsoa_metrics = pd.DataFrame(lsoa_list)
lsoa_exp_metrics = pd.DataFrame(lsoa_exp_list)
lsoa_od_metrics = pd.DataFrame(lsoa_od_list)

index = metrics.keys()

lsoa_metrics['LSOA'] = index
lsoa_metrics = lsoa_metrics.set_index('LSOA')

lsoa_exp_metrics['LSOA'] = index
lsoa_exp_metrics = lsoa_exp_metrics.set_index('LSOA')

lsoa_od_metrics['LSOA'] = index
lsoa_od_metrics = lsoa_od_metrics.set_index('LSOA')
```

```python
lsoa_exp_metrics
```

```python
lsoa_exp_metrics['Increase in LTS 1 Centrality Percent'] = lsoa_exp_metrics['Increase in LTS 1 Centrality Percent'].replace([np.inf, -np.inf], np.nan)
lsoa_exp_metrics['Increase in LTS 2 Centrality Percent'] = lsoa_exp_metrics['Increase in LTS 2 Centrality Percent'].replace([np.inf, -np.inf], np.nan)
```

```python
# Drop rows with NaN values in the specified column
lsoa_exp_metrics = lsoa_exp_metrics.dropna(subset=['Increase in LTS 1 Centrality Percent'])
lsoa_exp_metrics = lsoa_exp_metrics.dropna(subset=['Increase in LTS 2 Centrality Percent'])
```

```python
lsoa_exp_metrics
```

```python
fig, ax = plt.subplots(1,2,figsize=(10, 5))

# Plot histograms
ax[0].hist(lsoa_exp_metrics['% Increase in LTS 1 and 2 %']*100, bins=30, alpha=0.5, label = 'Change in Percent')
ax[0].hist(lsoa_exp_metrics['% Increase in LTS 1 and 2 KM']*100, bins=30, alpha=0.5, label = 'Change in KM')
ax[0].legend()
ax[0].set_title('Change in LTS 1 and 2 between 2016 and 2021', fontsize = 10)
ax[0].set_xlabel('Percentage Change')
ax[0].set_ylabel('LSOA Count')

# Plot histograms
ax[1].hist(lsoa_exp_metrics['Increase in LTS 1 Centrality Percent']*100, bins=30, alpha=0.5, label='LTS 1')
ax[1].hist(lsoa_exp_metrics['Increase in LTS 2 Centrality Percent']*100, bins=30, alpha=0.5, label='LTS 2')
ax[1].legend()
ax[1].set_title('Change in Mean Centrality of LTS Roads Between 2016 and 2021', fontsize = 10)
ax[1].set_xlabel('Percentage Change')
ax[1].set_ylabel('LSOA Count')

plt.show()
```

```python
# Replace inf values with the maximum value of the column
max_value = lsoa_od_metrics['Best path LTS1/2 Proportion Increase Percent'].replace([np.inf, -np.inf], np.nan).max()
lsoa_od_metrics['Best path LTS1/2 Proportion Increase Percent'] = lsoa_od_metrics['Best path LTS1/2 Proportion Increase Percent'].replace([np.inf, -np.inf], max_value)
```

```python
# Replace inf values with the maximum value of the column
max_value = lsoa_od_metrics['Dangerous road penalty increase percent'].replace([np.inf, -np.inf], np.nan).max()
lsoa_od_metrics['Dangerous road penalty increase percent'] = lsoa_od_metrics['Dangerous road penalty increase percent'].replace([np.inf, -np.inf], max_value)
```

```python
lsoa_od_metrics.dropna(subset=['Dangerous road penalty increase percent'])['Dangerous road penalty increase percent'] * -100
```

```python
fig, ax = plt.subplots(figsize=(5, 3))
ax.hist(lsoa_od_metrics['Best path LTS1/2 Proportion Increase Percent']*100, bins=30, alpha=0.5)
ax.set_title('Change in LTS 1 and 2 Composition of Best Route to Work', fontsize = 10)
ax.set_xlabel('Percentage Change')
ax.set_ylabel('LSOA Count')
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(5, 3))
ax.hist(lsoa_od_metrics.dropna(subset=['Dangerous road penalty increase percent'])['Dangerous road penalty increase percent'], bins=30, alpha=0.5)
ax.set_title('Change in Dangerous Road Penalty to Work', fontsize = 10)
ax.set_xlabel('Penalty Change')
ax.set_ylabel('LSOA Count')
plt.tight_layout()
plt.show()
```
