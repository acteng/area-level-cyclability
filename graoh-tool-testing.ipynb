```python
import graph_tool.all as gt
import networkx as nx
import pyintergraph
import pickle
import pandas as pd
```

```python
edge_import = pd.read_csv('graph_edges_test.csv')
```

```python
#list of tuples; u,v,length, lts

edge_list = []

for i,e in edge_import.iterrows():
    edge_list.append(tuple([str(e['u']),str(e['v']),e['length'],e['LTS']]))
    
g = gt.Graph()

elength = g.new_ep("float")
elts = g.new_ep("int")

g.add_edge_list(edge_list, hashed = True,eprops=[elength, elts])
```

```python
vp, ep = gt.betweenness(g,weight=elength)
```

```python
print(elts)
```

```python
lts0 = []
lts1 = []
lts2 = []
lts3 = []
lts4 = []

for e in g.edges():
    if elts[e] == 0:
        lts0.append(ep[e])
    elif elts[e] == 1:
        lts1.append(ep[e])
    elif elts[e] == 2:
        lts2.append(ep[e])
    elif elts[e] == 3:
        lts3.append(ep[e])
    elif elts[e] == 4:
        lts4.append(ep[e])
```
